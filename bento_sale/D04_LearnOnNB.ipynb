```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn import metrics
from sklearn.model_selection import cross_validate
from sklearn.naive_bayes import GaussianNB # ガウシアン
```

```python
train_df = pd.read_csv("./C01_train_encoded.csv")
#train_df.info()
```

```python
train_df.head(10)
```

```python
train_data = train_df.values
xs = train_data[:, 181:1208]  # project_title, project_essay1
y = train_data[:, 1]  # 正解データ列のみ抽出
```

```python
xs.shape
```

```python
y.shape
```

```python
train_X = xs[:150000, :]
train_Y = y[:150000]

test_X = xs[150000:, :]
test_Y = y[150000:]
```

```python
train_Y[0:20].astype(int)
```

```python
train_Y[train_Y[0,] == 1].sum() / train_Y.size
```

```python
from sklearn.metrics import accuracy_score
clf = GaussianNB() # 正規分布を仮定したベイズ分類

# NG acc=0.84 全て1と判定
#from sklearn.naive_bayes import MultinomialNB
#clf = MultinomialNB(alpha=1.0, class_prior=None, fit_prior=True)

clf.fit(train_X, train_Y.astype(int)) # 学習をする

pred_Y = clf.predict(test_X)
accuracy = accuracy_score(test_Y.astype(int), pred_Y)
print("acc = {:.2f}".format(accuracy))
```

```python
pred_proba_Y = clf.predict_log_proba(test_X)
pred_proba_Y[100:300]
```

```python
pred_Y = clf.predict(train_X)

accuracy = accuracy_score(train_Y.astype(int), pred_Y)
print("acc = {:.2f}".format(accuracy))
```
